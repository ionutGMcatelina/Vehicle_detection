```python
import numpy as np
import pickle
import pandas as pd
import random
import csv
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
```

```python
with open('classifier.pkl', 'rb') as fid:
    clf = pickle.load(fid)
```

```python
vehicles = pd.read_csv('vehiclesToDetect.csv')
```

```python
vehicles
```

```python
v = []
for i in range(0, len(vehicles.index)):
    v.append(vehicles.iloc[i])
```

```python
res = clf.predict(v)
```

```python
res
```

```python
with open('vehiclesDetected.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(res)
```
