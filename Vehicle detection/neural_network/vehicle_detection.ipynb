```python
import numpy as np
import pickle
import pandas as pd
import random
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
```

```python
input_file = "shuffle_vehicles.csv"
data = pd.read_csv(input_file, sep = None, engine = 'python')
```

```python
data
```

```python
clf = MLPClassifier(solver = 'adam',
                    alpha = 1e-5,
                    tol = 1e-5,
                    learning_rate_init = 0.001,
                    hidden_layer_sizes = (128, 64, 32),
                    random_state = 1,
                    verbose = True,
                    shuffle = True,
                    max_iter = 2000)
```

```python
values = data[data.columns.drop('vehicle')]
targets = data['vehicle']
```

```python
total = 8470
training = int(total * 0.85)
test = int(total * 0.15)

train_data = values[:training]
train_targets = targets[:training]
test_data = values[training:]
test_targets = targets[training:]
```

```python
clf.fit(train_data, train_targets)
```

```python
clf.score(test_data, test_targets)
```

```python
with open('classifier.pkl', 'wb') as fid:
    pickle.dump(clf, fid) 
```
